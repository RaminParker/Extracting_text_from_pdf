# Merging PDF files

[Function was taken from here](https://stackoverflow.com/questions/3444645/merge-pdf-files)

```
pdfs = ['Handelsblatt_1.pdf', 'BaFin.pdf']

merger = PdfFileMerger()

for pdf in pdfs:
    merger.append(open(pdf, 'rb'))

with open('collection_pdfs.pdf', 'wb') as fout:
    merger.write(fout)
```

# Extracting text from pdf (single page)

Open file in binary mode and saved the file object as pdfFileObj.

```
pdfFileObj = open('Handelsblatt_1.pdf', 'rb') 
```

Create an object of PdfFileReader class of PyPDF2 module and  pass the pdf file object & get a pdf reader object.

```
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

Number of pages in pdf

```
numPages=pdfReader.numPages
numPages
```

```
type(numPages)
```

Now, we create an object of PageObject class of PyPDF2 module. pdf reader object has function getPage() which takes page number (starting form index 0) as argument and returns the page object.

```
pageObj = pdfReader.getPage(0)
```

Extracting text from page.

This will print the text.

```
#print(pageObj.extractText())
```

# Extract text from sequetial pages

```
page_content=""                # define variable for using in loop.
for page_number in range(numPages):
    page = pdfReader.getPage(page_number)
    page_content += page.extractText()     # concate reading pages.
```
